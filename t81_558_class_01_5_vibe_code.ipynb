# Part 1.5: Vibe Coding: Acceptable AI Use in this Course

This module introduces modern approaches to **vibe coding** and **code generation** using generative AI. These tools allow you to move from ideas to working code quickly, which is especially useful when building and experimenting with deep learning models in PyTorch. While both concepts rely on large language models, they operate at different levels of abstraction. *Code generation* focuses on producing functions, classes, or scripts from prompts, often within a traditional development workflow. *Vibe coding* goes further by allowing you to describe an entire application or system in natural language and have the platform assemble, run, and sometimes deploy it.

In this course, you will use both approaches. You will generate PyTorch code from structured prompts, refine it through debugging, and learn how to guide models toward correct and efficient implementations. Prompting plays a central role. Small changes in wording, constraints, or examples can significantly affect the output. You will also practice debugging AI-generated code, which is an essential skill. These systems can produce correct-looking code that contains subtle errors, so verification and iteration are part of the workflow.

## Course Policy for Vibe Coding and Code Generation

In this course, the use of AI tools for code generation is **expected and encouraged**. You will often be provided with prompts designed to help you build specific PyTorch components such as neural networks, training loops, and data pipelines.

The goal is not to avoid AI assistance. The goal is to learn how to use it effectively and responsibly.

- You may use AI tools to generate and refine code for assignments  
- You should understand and be able to explain any code you submit  
- You will be expected to debug and correct generated code when it is wrong  
- Prompts provided in the course are designed to guide you toward correct implementations  

You should treat AI as a collaborator. It can accelerate development, but it does not replace understanding. The strongest outcomes come from combining generated code with your own reasoning and validation.

## Code Generation with ChatGPT

ChatGPT is a general-purpose conversational interface to large language models. It is often the starting point for code generation because it allows flexible interaction, iterative refinement, and explanation.

In this course, ChatGPT is useful for generating PyTorch code, explaining model behavior, and debugging errors.

- Works through natural language prompts and follow-up questions  
- Can generate complete code snippets, including neural network definitions and training loops  
- Useful for explaining concepts such as backpropagation or tensor shapes  
- Supports iterative refinement through conversation  

ChatGPT has both free and paid tiers. The paid versions provide access to more capable models and higher usage limits, which can be helpful for larger or more complex tasks.

When using ChatGPT, focus on clear prompts. Specify the framework (PyTorch), include constraints, and ask for explanations when needed. This improves both correctness and learning.

## Code Generation with GitHub Copilot

GitHub Copilot is an AI coding assistant integrated directly into development environments such as VS Code and JetBrains IDEs. It provides inline suggestions as you write code.

In contrast to ChatGPT, Copilot operates within your existing workflow and reacts to the code you are currently editing.

- Provides real-time autocomplete for functions, classes, and boilerplate  
- Learns from the surrounding context in your file  
- Can generate tests, documentation, and simple model code  
- Works well for filling in repetitive or standard patterns  

Copilot is a paid tool, though it is often free for students or available through institutional licenses.

For this course, Copilot is most useful when implementing known patterns such as forward passes, loss calculations, and dataset handling. You should still review its suggestions carefully, especially for correctness in tensor operations.

## Code Generation with Cursor

Cursor is an AI-first code editor designed to integrate large language models deeply into the development process. It extends the idea of Copilot by allowing multi-file reasoning and structured edits across a project.

Cursor is particularly useful when working with larger codebases or when modifying multiple components of a deep learning system.

- Can analyze and modify entire repositories, not just a single file  
- Supports natural language instructions for refactoring or adding features  
- Shows diffs so you can review changes before applying them  
- Maintains awareness of project structure and dependencies  

Cursor is a paid product with a free tier. The paid version provides more usage and access to stronger models.

In this course, Cursor is helpful when your PyTorch projects grow beyond a single script. It allows you to update models, training code, and evaluation logic in a coordinated way.

## Code Generation with Replit

Replit is an online development environment that includes AI-assisted coding and deployment features. It is one of the easiest platforms for getting started with vibe coding because it requires no local setup.

Replit allows you to describe an application and have it generate a working environment, including backend and frontend components.

- Runs entirely in the browser with no installation required  
- Can generate full applications from prompts  
- Provides hosting and deployment directly within the platform  
- Includes collaborative features for sharing and editing projects  

Replit offers a free tier with usage limits, along with paid plans for more compute and persistent resources.

For this course, Replit is useful for rapid experimentation and for understanding how high-level descriptions translate into working systems. It is less focused on fine-grained control and more on speed and convenience.

## Code Generation with Claude Code

Claude Code represents a newer category of tools often called **agentic coding systems**. These systems can plan and execute multi-step coding tasks with limited supervision.

Rather than responding to a single prompt, Claude Code can iterate through a sequence of actions such as writing code, running it, detecting errors, and applying fixes.

- Handles multi-step workflows including debugging and refactoring  
- Can operate across multiple files in a project  
- Supports higher-level instructions such as “implement and test this feature”  
- Often used through command-line or integrated development tools  

Claude Code is typically part of a paid ecosystem, with pricing based on usage of underlying models.

In this course, tools like Claude Code illustrate where code generation is heading. They are useful for complex tasks, but they also require careful oversight. You should always review changes, verify outputs, and ensure that the generated code matches the intended design.

These tools represent different points along a spectrum, from simple code suggestion to full system generation. You will use them throughout the course to accelerate development, explore ideas, and deepen your understanding of deep learning systems.

The key skill is not just generating code, but **guiding, evaluating, and correcting it**.
