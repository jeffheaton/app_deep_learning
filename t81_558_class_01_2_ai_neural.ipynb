# Part 1.2: AI and Neural Networks

Artificial Intelligence (AI) denotes the class of computational systems that perform tasks requiring decision-making, pattern recognition, or adaptation under uncertainty. The term is defined functionally rather than by implementation. Multiple paradigms fall under this definition, including symbolic methods, statistical models, and neural networks. The boundaries between subfields are nested and partially overlapping. The diagram situates commonly used terms within this hierarchy.

![AI Hierarchy Diagram](https://s3.us-east-1.amazonaws.com/data.heatonresearch.com/images/jupyter/ai_levels.png)

- **Artificial Intelligence**
  - The outermost set. Includes all computational approaches to intelligent behavior.
  - Encompasses rule-based systems, search, optimization, probabilistic inference, and learning-based methods.
  - No assumption of learning is required.

- **Machine Learning**
  - Subset of AI defined by parameter estimation from data.
  - Models are trained rather than explicitly programmed.
  - Includes linear models, decision trees, kernel methods, and neural networks.

- **Deep Learning**
  - Subset of Machine Learning based on multi-layer neural networks.
  - Computation is organized as compositions of parameterized functions.
  - Training is typically performed with gradient-based optimization over large datasets.

- **Generative AI**
  - Subset of Deep Learning focused on modeling data distributions.
  - Produces new samples $x \sim p_{\theta}(x)$ rather than predicting labels.
  - Includes large language models, diffusion models, and generative adversarial networks.

- **Predictive AI**
  - Overlapping subset of Machine Learning.
  - Focuses on estimating conditional distributions $p(y \mid x)$.
  - Includes classification, regression, and forecasting tasks.
  - Not restricted to neural networks; may use any supervised learning method.

## What is Artificial General Intelligence (AGI)

Artificial General Intelligence (AGI) denotes a system that can perform a broad range of cognitive tasks across domains at a level comparable to a human. The defining property is transfer: competence acquired in one domain applies to others without retraining for each task. Current systems do not satisfy this condition. They are optimized for specific objective functions and operate within narrow distributions.

![AI Hierarchy Diagram](https://s3.us-east-1.amazonaws.com/data.heatonresearch.com/images/jupyter/on_trick.png)

The image presents several dominant model classes used in modern AI systems. Each corresponds to a distinct problem formulation and training objective. These systems do not share a unified representation that allows arbitrary transfer between tasks.

- **Transformer**
  - Sequence modeling architecture based on attention mechanisms.
  - Used for language, code, and other structured sequential data.
  - Optimized for next-token prediction or similar objectives.

- **Diffusion**
  - Generative model based on iterative denoising processes.
  - Produces images, audio, and other continuous signals.
  - Training objective differs from sequence prediction.

- **RL Search**
  - Reinforcement learning combined with search procedures.
  - Used in domains such as games and planning.
  - Requires explicit reward structures and environment interaction.

- **Perception**
  - Models for vision, speech, and sensor interpretation.
  - Typically trained on labeled datasets for classification or detection.
  - Focused on mapping raw input to structured representations.

These systems do not substitute for one another. A transformer-based language model does not perform image synthesis without architectural modification. A diffusion model does not execute search over game states. Each is a specialized solution.

This specialization is observable in practice. Large language models such as ChatGPT generate coherent explanations and code but perform poorly at precise combinatorial tasks such as chess. Dedicated engines such as Stockfish evaluate positions and compute optimal moves with high accuracy but do not produce natural language explanations or general reasoning. The contrast is sharp: an Atari 2600-era chess implementation, given fixed rules and deterministic evaluation, can outperform a general-purpose language model in gameplay. The model trained for language lacks the search depth and evaluation function required for strong play.

The following diagram situates current systems within a broader progression of capability.

![AI Hierarchy Diagram](https://s3.us-east-1.amazonaws.com/data.heatonresearch.com/images/jupyter/levels_of_ai.png)

The stages in the diagram represent increasing generality and autonomy. Each transition requires changes in representation, learning, and control.

- **Narrow / Specialized AI**
  - Systems optimized for a single task or domain.
  - Performance depends on alignment between training data and deployment conditions.
  - No transfer outside the trained objective.

- **Multi-task / Generalist AI**
  - Systems trained on multiple tasks with shared parameters.
  - Limited transfer between related tasks.
  - Performance degrades as tasks diverge.

- **Artificial General Intelligence (AGI)**
  - Unified system capable of cross-domain reasoning and learning.
  - Adapts to new tasks without task-specific retraining.
  - Requires general representations and flexible control mechanisms.

- **Recursive Self-Improvement (RSI)**
  - System modifies its own architecture or training process.
  - Improvement loop depends on stability and evaluation criteria.
  - Not demonstrated in current systems.

- **Superintelligence**
  - Hypothetical stage beyond human-level performance across domains.
  - Requires sustained improvement beyond AGI.
  - No empirical examples.

Movement between these stages is not incremental scaling alone. Increased parameter count or training data improves performance within a regime but does not establish generality. Transitions require new mechanisms for abstraction, memory, and task composition.

The final diagram frames the relationship between human capability and machine assistance.

![AI Hierarchy Diagram](https://s3.us-east-1.amazonaws.com/data.heatonresearch.com/images/jupyter/ai_dominance.jpg)

The progression shows increasing integration of automated systems into decision processes. The intermediate phases involve collaboration rather than replacement.

- **Manual Era**
  - Human performs all tasks directly.
  - Tools provide mechanical assistance only.

- **Narrow AI (rules)**
  - Deterministic systems encode explicit procedures.
  - Limited adaptability.

- **Assistant AI**
  - Systems provide recommendations or partial outputs.
  - Human retains control over decisions.

- **Centaur Phase**
  - Human and AI operate jointly on the same task.
  - Performance exceeds either alone in structured domains.

- **Autonomous Mastery**
  - System executes tasks without human intervention.
  - Requires reliability across varied conditions.

Current systems reside between Narrow AI and the early Centaur phase. They augment human capability but do not replace general reasoning. AGI would require a system that unifies the capabilities shown in the first diagram and operates across the stages shown in the second without task-specific reconfiguration.

# Human Brains and AI Are Very Different

Human cognition and contemporary AI systems arise from distinct physical substrates and computational models. The brain is a biological network of neurons connected by synapses with time-varying strengths; signaling is sparse, asynchronous, and modulated by chemical processes. Modern AI systems are implemented on digital hardware; computation proceeds through synchronous execution of arithmetic operations on fixed-precision representations. Training procedures differ in mechanism and objective. Biological learning is continuous, local, and energy-constrained; machine learning relies on global optimization (e.g., stochastic gradient descent) applied to large parameterized functions. These differences constrain achievable behaviors, scaling strategies, and failure modes, and they persist under improvements in hardware or algorithms.

The table presents a comparison between a human brain and a contemporary data-center GPU (NVIDIA H100). Each row isolates a measurable property: power consumption, physical scale, count of computational elements, manufacturing process, operational lifetime, fault tolerance, and deployment configuration. The quantities are drawn from standard approximations. Synapse counts for the brain are reported as order-of-magnitude estimates; parameter counts for the GPU correspond to a typical large model footprint under reduced-precision formats. Parallelism is described at the level of execution units and signaling events rather than instruction sets. The entries are not normalized to a common computational model; no direct equivalence is assumed between synapses and parameters, or between neural signaling and SIMD execution.

| Feature          | Human Brain                              | H100 GPU (typical server config)                                      |
|------------------|------------------------------------------|------------------------------------------------------------------------|
| Power            | ~20 W                                    | 350–700 W                                                              |
| Mass             | ~1.4 kg                                  | ~1.7 kg                                                                |
| Volume           | ~1.4 L                                   | ~1.1 L                                                                 |
| Compute Elements | ~10¹⁴ synapses (dynamic, analog)         | ~50B parameters (FP16/BF16, static)                                   |
| Parallelism      | Massive, event-driven                    | ~10⁴–10⁵ parallel threads (SIMT)                                      |
| Manufacturing    | Biological development (~9 months)       | Global supply chain, advanced lithography, precision assembly         |
| Lifespan         | ~80 years                                | ~3–5 years (practical lifecycle)                                      |
| Fault Tolerance  | High (neuroplasticity, redundancy)       | Low (hardware faults → replace)                                       |
| Scaling Unit     | 1 brain                                  | 8+ GPUs per node                                                       |

The comparison exhibits distinct design regimes. The brain operates under strict power constraints, maintains functionality under component failure, and adapts through continuous structural modification. The GPU operates under high power budgets, relies on exact component behavior, and scales through replication and interconnect bandwidth. Manufacturing constraints differ in both process and time scale. The brain develops through biological growth with embedded repair mechanisms; the GPU is fabricated through multi-stage lithographic processes with no intrinsic recovery from defects. The observed differences follow from these constraints and are not reducible to a single efficiency metric.
