# Part 1.1: Overview

Deep learning is a group of exciting new technologies for neural networks. [[Cite:lecun2015deep]](https://www.nature.com/articles/nature14539) By using a combination of advanced training techniques neural network architectural components, it is now possible to train neural networks of much greater complexity. This book introduces the reader to deep neural networks, regularization units (ReLU), convolution neural networks, and recurrent neural networks. High-performance computing (HPC) aspects demonstrate how deep learning can be leveraged both on graphical processing units (GPUs), as well as grids. Deep learning allows a model to learn hierarchies of information in a way that is similar to the function of the human brain. The focus is primarily upon the application of deep learning, with some introduction to the mathematical foundations of deep learning. Readers will make use of the Python programming language to architect a deep learning model for several real-world data sets and interpret the results of these networks. [[Cite:goodfellow2016deep]](https://www.deeplearningbook.org/)

Generative AI (GenAI) systems have become among the most widely deployed applications of deep learning since approximately 2020. Large language models, diffusion models, and multimodal architectures now account for a substantial fraction of both research output and production systems. The core machinery is not new: transformer architectures, gradient descent, and backpropagation all predate this period. What changed was scale. Parameter counts, training corpus sizes, and compute budgets increased by several orders of magnitude, and this scaling produced qualitative improvements in capability that smaller models did not exhibit.

Generative models are implemented as deep neural networks trained to approximate a distribution over structured outputs. A large language model is a transformer trained to model $P(x_t | x_1, ..., x_{t-1})$ over tokens; at inference time, outputs are drawn autoregressively from this distribution. A diffusion model learns to reverse a Markov chain that progressively corrupts data with Gaussian noise, parameterizing the reverse transitions with a deep network trained via a variational lower bound on the data log-likelihood. In both cases, the network consists of many layers of parameterized, nonlinear transformations trained end-to-end on large datasets.

Several problem classes that previously required purpose-built architectures are now handled by general-purpose generative models. Sequence-to-sequence NLP tasks, including machine translation, summarization, and named entity recognition, were historically addressed with task-specific models such as attention-based encoder-decoders and fine-tuned BERT variants; these have been largely displaced by prompted or fine-tuned LLMs. Image generation, formerly the domain of GANs trained on narrow distributions, is now handled primarily by diffusion models and autoregressive vision transformers. Speech recognition and synthesis, once built on HMM-DNN hybrids, are increasingly unified under sequence models in the LLM or diffusion family. Settings with hard output constraints or very limited training data still favor specialized methods, but the generative paradigm has absorbed a large fraction of what was once a diverse ecosystem of purpose-built architectures.

## What is Deep Learning

The focus of this book is deep learning, a prevalent type of machine learning that builds on the original neural networks popularized in the 1980s. There is very little difference between how a deep neural network is trained and how the first neural network is trained. We've always been able to create and calculate deep neural networks. A deep neural network is nothing more than a neural network with many layers. While we've always been able to create/calculate deep neural networks, we've lacked an effective means of training them. Deep learning provides an efficient means to train deep neural networks.

If deep learning is a type of machine learning, then the question arises: "What is machine learning?" Figure 1.ML-DEV illustrates how machine learning differs from traditional software development.

**Figure 1.ML-DEV: ML vs Traditional Software Development**
![ML vs Traditional Software Development](https://s3.us-east-1.amazonaws.com/data.heatonresearch.com/images/jupyter/dl-domains-2.png "Machine Learning vs Traditional Software Development")

* **Traditional Software Development** - Programmers create programs that specify how to transform input into the desired output. Humans explicitly encode the rules, and computers execute them deterministically.

* **Machine Learning** - Programmers create models that can learn to produce the desired output for given input. This learning fills the traditional role of the computer program. Instead of writing rules, humans provide examples of inputs paired with desired outputs, and the computer derives the rules (the model) through training.

* **Generative AI (Inference)** - Rather than training a new model for each task, programmers and end users interact with a pre-trained foundation model by providing prompts. The model applies its learned rules to generate novel content such as text, images, or audio. The human role shifts from writing code to specifying intent in natural language.

* **Generative AI for Code Generation** - The model's output is program code that is executed by a traditional computer to produce the final result. This chains generative AI with classical execution: the human expresses intent, the model synthesizes the rules, and the computer runs them. In agentic workflows, execution results and errors feed back into the model, allowing it to revise and re-run the code with humans providing oversight at designated checkpoints rather than writing each line.

Researchers have applied machine learning to many different areas.  This class explores three specific domains for the application of deep neural networks, as illustrated in Figure 1.ML-DOM.

**Figure 1.ML-DOM: Application of Machine Learning**
![Application of Machine Learning](https://data.heatonresearch.com/images/jupyter/dl-domains.jpg "Application of Machine Learning")

* **Computer Vision** - Learning from visual data: classifying images, detecting objects, segmenting regions. A classifier trained on labeled images learns filters that activate on edges, textures, and higher-level features across successive convolutional layers.
* **Tabular Data** - Fully connected networks trained on fixed-width feature vectors, where each input dimension corresponds to a named measurement. The iris dataset, which uses four floral measurements to predict species, is a canonical example.
* **Natural Language Processing (NLP)** - Sequence modeling over tokenized text. Transformer-based models process input sequences via self-attention and are trained on objectives such as next-token prediction or masked token recovery; the resulting representations support classification, translation, summarization, and generation.
* **Reinforcement Learning** - A neural network parameterizes a policy that maps states to actions. Training maximizes cumulative reward through environmental interaction, using algorithms such as policy gradient methods or Q-learning.
* **Time Series** - Sequence modeling over temporally ordered observations. Applications include financial forecasting, anomaly detection, and speech recognition. NLP is a special case when the sequence consists of discrete tokens.
* **Generative Models** - Networks trained to model a data distribution and sample from it, producing synthetic outputs such as images or text. This text covers the deep learning foundations of large language models and image generation specifically.

## Regression, Classification and Beyond

Machine learning problems are broadly categorized as supervised or unsupervised:

* **Supervised learning** - Each training example carries a known label or target value. The network is trained to minimize error against these labels. Subdivides into classification, where the output is a probability distribution over a discrete label set, and regression, where the output is one or more continuous values. The same architecture can produce both simultaneously, with the output layer partitioned into classification and regression heads trained under separate loss terms.
* **Unsupervised learning** - No labels are provided; the model learns structure from the data alone. Clustering, dimensionality reduction, and density estimation are typical objectives.

Deep learning extends this space with two additional training regimes:

* **Reinforcement learning** - A network is trained via interaction with an environment, optimizing a cumulative reward signal rather than a fixed labeled dataset.
* **Adversarial learning** - Two networks are trained in opposition. A generator produces synthetic data; a discriminator distinguishes synthetic from real. Each network's loss is defined in terms of the other's output.

Neural network inputs and outputs are not restricted to a single data type. A network may take an image, a token sequence, a fixed-width feature vector, or a combination as input, and may produce a class label, a continuous score, a token sequence, or an image as output. The choice of representation determines the loss function and output layer structure, but not necessarily the internal architecture.

## Why Deep Learning?

For tabular data, neural networks do not consistently outperform other model classes, including:

* [Support Vector Machines](https://en.wikipedia.org/wiki/Support-vector_machine)
* [Random Forests](https://en.wikipedia.org/wiki/Random_forest)
* [Gradient Boosted Machines](https://en.wikipedia.org/wiki/Gradient_boosting)

On low-dimensional tabular tasks, gradient boosted trees in particular frequently match or exceed deep network performance with less compute and less sensitivity to hyperparameter choice. For images, video, text, and audio, deep neural networks dominate; no other model class is competitive at scale on these modalities.
