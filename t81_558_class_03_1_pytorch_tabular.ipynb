# Part 3.1: Should Neural Networks be Used for Tabular Data

Neural networks are a powerful and flexible modeling approach, but they are not always the best choice for tabular data. In many real-world tabular problems, such as structured business data, insurance records, or financial datasets, models like gradient boosting machines often outperform neural networks while requiring less tuning and computational effort. These models are designed to handle mixed feature types, missing values, and complex interactions that are common in tabular data. As a result, they frequently provide stronger baseline performance and are easier to use in practice.

Neural networks tend to perform best in domains with spatial, sequential, or unstructured data, such as images, audio, and natural language. When applied to tabular data, they usually require careful preprocessing, feature scaling, and architecture design to reach similar levels of performance. Without this additional work, a neural network can underperform simpler and more specialized approaches, especially when working with smaller datasets.

Even with these limitations, it is important to learn how neural networks apply to tabular data. Neural networks provide a consistent framework that can extend across many data types, making them useful for combining structured data with text, images, or time series. Techniques such as categorical embeddings and attention mechanisms are also improving neural networks' ability to handle structured inputs. In addition, learning neural networks deepens understanding of optimization, generalization, and model behavior, which carries over to many other areas of artificial intelligence.

For these reasons, neural networks remain an important part of a complete machine learning toolkit, even when they are not always the top choice for tabular problems.

## Missing Values

Missing values are a reality of machine learning. Ideally, every row of data will have values for all columns. However, this is rarely the case. Most of the values are present in the MPG database. However, there are missing values in the horsepower column. A common practice is to replace missing values with the column's median. The program calculates the [median](https://en.wikipedia.org/wiki/Median). The following code replaces any NA values in horsepower with the median:


```python
df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)
print(f"horsepower has na? {pd.isnull(df['horsepower']).values.any()}")

print("Filling missing values...")
med = df["horsepower"].median()
df["horsepower"] = df["horsepower"].fillna(med)
# df = df.dropna() # you can also simply drop NA values

print(f"horsepower has na? {pd.isnull(df['horsepower']).values.any()}")
```

## Dealing with Outliers

Outliers are values that are unusually high or low. We typically consider outliers to be values several standard deviations from the mean. Sometimes outliers are simply errors; they result from [observation error](https://en.wikipedia.org/wiki/Observational_error). Outliers can also be truly large or small values that may be difficult to address. The following function can remove such values.


```python
# Remove all rows where the specified column is +/- sd standard deviations
def remove_outliers(df, name, sd):
    drop_rows = df.index[(np.abs(df[name] - df[name].mean())
                          >= (sd * df[name].std()))]
    df.drop(drop_rows, axis=0, inplace=True)
```

The code below will drop every row from the Auto MPG dataset where the horsepower is two standard deviations or more above or below the mean.


```python
import numpy as np


df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

# create feature vector
med = df["horsepower"].median()
df["horsepower"] = df["horsepower"].fillna(med)

# Drop the name column
df.drop(columns=["name"], inplace=True)

# Drop outliers in horsepower
print("Length before MPG outliers dropped: {}".format(len(df)))
remove_outliers(df, "mpg", 2)
print("Length after MPG outliers dropped: {}".format(len(df)))

pd.set_option("display.max_columns", 0)
pd.set_option("display.max_rows", 5)
display(df)
```

## Training and Validation

We must evaluate a machine learning model based on its ability to predict values that it has never seen before. Because of this, we often divide the training data into a validation set and a training set. The machine learning model will learn from the training data, but ultimately be evaluated based on the validation data.

- **Training Data** - **In Sample Data** - The data that the neural network used to train.
- **Validation Data** - **Out of Sample Data** - The data that the machine learning model is evaluated upon after it is fit to the training data.

There are two effective means of dealing with training and validation data:

- **Training/Validation Split** - The program splits the data according to some ratio between a training and validation (hold-out) set. Typical rates are 80% training and 20% validation.
- **K-Fold Cross Validation** - The program splits the data into several folds and models. Because the program creates the same number of models as folds, the program can generate out-of-sample predictions for the entire dataset.

The code below splits the MPG data into a training and validation set. The training set uses 80% of the data, and the validation set uses 20%. Figure 2.TRN-VAL shows how we train a model on 80% of the data and then validate against the remaining 20%.

**Figure 2.TRN-VAL: Training and Validation**
![Training and Validation](https://raw.githubusercontent.com/jeffheaton/t81_558_deep_learning/master/images/class_1_train_val.png "Training and Validation")


```python
df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

# Usually a good idea to shuffle
df = df.reindex(np.random.permutation(df.index))

mask = np.random.rand(len(df)) < 0.8
trainDF = pd.DataFrame(df[mask])
validationDF = pd.DataFrame(df[~mask])

print(f"Training DF: {len(trainDF)}")
print(f"Validation DF: {len(validationDF)}")
```

## Saving a Dataframe to Pickle

A variety of software programs can use CSV-formatted text files. However, they take longer to generate and can sometimes lose a small amount of precision during conversion. Generally, you will output to CSV because it is very compatible, even outside of Python. Another format is [Pickle](https://docs.python.org/3/library/pickle.html). The code below saves the DataFrame to a Pickle file. Pickle stores data in the exact binary representation used by Python. The benefit is that there is no data loss when converting to CSV format. The disadvantage is that, in general, only Python programs can read Pickle files.


```python
import os
import pickle

path = "."

df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

filename_write = os.path.join(path, "auto-mpg-shuffle.pkl")
df = df.reindex(np.random.permutation(df.index))

with open(filename_write, "wb") as fp:
    pickle.dump(df, fp)
```

Loading the pickle file back into memory is accomplished by the following lines of code. Notice that the index numbers are still jumbled from the previous shuffle? Loading the CSV rebuilt (in the last step) did not preserve these values.


```python
import os
import pickle

path = "."

df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/auto-mpg.csv", na_values=["NA", "?"]
)

filename_read = os.path.join(path, "auto-mpg-shuffle.pkl")

with open(filename_read, "rb") as fp:
    df = pickle.load(fp)

pd.set_option("display.max_columns", 7)
pd.set_option("display.max_rows", 5)
display(df)
```
