# Part 3.1: Deep Learning and Neural Network Introduction

A neural network is a parameterized function composed of alternating linear transformations and pointwise nonlinearities. The "deep" in deep learning refers to networks with many such composed layers — depth is the single most important architectural factor distinguishing modern networks from their predecessors, enabling hierarchical representations where early layers detect low-level structure and later layers compose those detections into progressively abstract features.

Neural networks have undergone two cycles of diminished interest followed by resurgence. The current period, beginning around 2012, has proven far more durable. This course covers only contemporary methods: modern architectures, current training procedures, and the theoretical understanding that has developed alongside them.

## Input Representations

The input to a neural network is a tensor of fixed shape. Historically this was a one-dimensional feature vector — a flat array of numeric values analogous to a row in a spreadsheet — and changing its length required retraining the network from scratch. Modern architectures accept inputs of arbitrary rank:

* **1D vector**: tabular data, encoded audio, or any flat feature representation
* **2D matrix**: grayscale images, spectrograms
* **3D tensor**: color images (height × width × channels), short video clips
* **nD tensor**: higher-order inputs to convolutional or volumetric models

Before convolutional neural networks (CNNs), images were either encoded into an intermediate 1D form or flattened by stacking pixel rows into a single array. CNNs eliminate this by operating directly on the spatial structure of the input, passing the matrix through layers that preserve locality. A 64×64 grayscale image fed to a CNN has a 2D input of shape 64×64; the same image fed to a classical fully connected network would be flattened to a 4,096-dimensional vector. Both descriptions are correct — they refer to different senses of "dimension."

This ambiguity in terminology is worth making precise. When discussing a 1D input vector, dimension refers to the number of elements in the array: a network with ten input neurons takes a 10-dimensional vector. When discussing tensor rank, dimension refers to the number of axes: a color image is a 3D tensor regardless of its spatial resolution. Context usually disambiguates these uses, but the distinction matters when reasoning about architecture.

This book begins with 1D input and fully connected networks. Later modules introduce CNNs and higher-dimensional inputs, followed by the transformer architecture, which processes sequences of vectors and now dominates language modeling, vision, and multimodal tasks. The final modules cover generative models — including diffusion models and the architecture underlying Stable Diffusion — and the theory of embeddings and representation learning that connects these systems.

## Classification or Regression

Like many models, neural networks can function in classification or regression:

- **Regression** - You expect a numerical value as your neural network’s prediction.
- **Classification** - You expect a class/category as your neural network's prediction.

A classification and regression neural network is shown by Figure 3.CLS-REG.

**Figure 3.CLS-REG: Neural Network Classification and Regression**
![Neural Network Classification and Regression](https://data.heatonresearch.com/images/wustl/class/class_2_ann_class_reg.png "Neural Network Classification and Regression")

Notice that the regression neural network output is numeric, while the classification output is a class. Regression, or two-class classification, networks always have a single output. Neural network classifiers have an output neuron for each category.

## Neurons and Layers

Most neural network structures use some type of neuron. Many different neural networks exist, and programmers introduce experimental neural network structures. Consequently, it is not possible to cover every neural network architecture. However, there are some commonalities among neural network implementations. A neural network algorithm typically consists of individual, interconnected units, even though these units may or may not be called neurons. The name for a neural network processing unit varies across the literature. It could be called a node, neuron, or unit.

A diagram shows the abstract structure of a single artificial neuron in Figure 3.ANN.

**Figure 3.ANN: An Artificial Neuron**
![An Artificial Neuron](https://data.heatonresearch.com/images/wustl/class/class_2_abstract_nn.png "An Artificial Neuron")


The artificial neuron receives input from one or more sources, which may be other neurons or data fed into the network by a computer program. This input is usually floating-point or binary. Often, binary input is encoded to floating-point by representing true or false as 1 or 0. Sometimes the program also represents binary information using a bipolar system, with true as 1 and false as -1.

An artificial neuron multiplies each of these inputs by a weight. Then it adds these multiplications and passes this sum to an activation function. Some neural networks do not use an activation function. The following equation summarizes the calculated output of a neuron:

$$ f(x,w) = \phi(\sum_i(w_i \cdot x_i)) $$

In the above equation, the variables $x$ and $w$ represent the input and weights of the neuron. The variable $i$ corresponds to the number of weights and inputs. You must always have the same number of weights as inputs. The neural network multiplies each weight by its respective input and feeds the products of these multiplications into an activation function, denoted by the Greek letter $\phi$ (phi). This process results in a single output from the neuron.

## Types of Neurons

In the last section, we briefly introduced the idea that there are different types of neurons. Not every neural network will use every kind of neuron. It is also possible for a single neuron to fill the role of several different neuron types. Now we will explain all the neuron types described in the course.

There are usually four types of neurons in a neural network:

* **Input Neurons** - We map each input neuron to one element in the feature vector.
* **Hidden Neurons** - Hidden neurons allow the neural network to be abstract and to process input into output.
* **Output Neurons** - Each output neuron calculates one part of the output.
* **Bias Neurons** - Work similarly to the y-intercept of a linear equation.

We place each neuron into a layer:

* **Input Layer** - The input layer accepts feature vectors from the dataset. Input layers usually have a bias neuron.
* **Output Layer** - The output from the neural network. The output layer does not have a bias neuron.
* **Hidden Layers** - Layers between the input and output layers. Each hidden layer typically includes a bias neuron.

## Input and Output Neurons

Nearly every neural network has input and output neurons. The input neurons receive data from the program. The output neuron provides processed network data back to the program. The program will group these input and output neurons into separate layers called the input and output layers. The program normally represents the input to a neural network as an array or vector. The number of elements contained in the vector must equal the number of input neurons. For example, a neural network with three input neurons might accept the following input vector:

$$ [0.5, 0.75, 0.2] $$

Neural networks typically accept floating-point vectors as their input. To be consistent, we will represent the output of a single-output neuron network as a single-element vector. Likewise, neural networks will output a vector with a length equal to the number of output neurons. The output will often be a single value from a single output neuron.

## Hidden Neurons

Hidden neurons have two essential characteristics. First, hidden neurons receive input only from other neurons, such as other hidden neurons or the input layer. Second, hidden neurons only output to other neurons, such as other hidden neurons or the output layer. Hidden neurons help the neural network understand the input and form the output. Programmers often group hidden neurons into fully connected hidden layers. However, these hidden layers do not directly process the incoming data or the eventual output.


A common question for programmers concerns the number of hidden neurons in a network. Since the answer to this question is complex, more than one section of the course will include a relevant discussion of the number of hidden neurons. Before deep learning, researchers generally believed that anything beyond a single hidden layer was excessive. [[Cite:hornik1989multilayer]](https://www.sciencedirect.com/science/article/abs/pii/0893608089900208) Researchers have proven that a single-hidden-layer neural network can function as a universal approximator. In other words, this network should be able to learn to produce (or approximate) any output from any input as long as it has enough hidden neurons in a single layer.


Training refers to the process that determines good weight values. Before the advent of deep learning, researchers feared that additional layers would lengthen training time or lead to overfitting. Both concerns are valid; however, increased hardware speeds and clever techniques can mitigate them. Before researchers introduced deep learning techniques, we did not have an efficient way to train a deep network, which is a neural network with many hidden layers. Although a single-hidden-layer neural network can theoretically learn anything, deep learning facilitates a more complex representation of patterns in the data.

## Bias Neurons

Programmers add bias neurons to neural networks to help them learn patterns. Bias neurons function like input neurons that always produce a value of 1. Because the bias neurons have a constant output of 1, they are not connected to the previous layer. The value 1, called the bias activation, can be set to any value other than 1. However, 1 is the most common bias activation. Not all neural networks have bias neurons. Figure 3.BIAS shows a single-hidden-layer neural network with bias neurons:

**Figure 3.BIAS: Neural Network with Bias Neurons**
![Neural Network with Bias Neurons](https://data.heatonresearch.com/images/wustl/class/class_2_ann.png "Neural Network with Bias Neurons")

The above network contains three bias neurons. Except for the output layer, every level includes a single bias neuron. Bias neurons allow the program to shift the output of an activation function. We will see precisely how this shifting occurs later in the module, when we discuss activation functions.

## Other Neuron Types

The individual units that comprise a neural network are not always called neurons. Researchers will sometimes refer to these neurons as nodes, units, or summations. You will almost always construct neural networks of weighted connections between these units.

## Why are Bias Neurons Needed?

The activation functions from the previous section specify the output of a single neuron. Together, a neuron's weight and bias shape its activation to produce the desired output. To see how this process occurs, consider the following equation. It represents a single-input sigmoid activation neural network.


$$ f(x,w,b) = \frac{1}{1 + e^{-(wx+b)}} $$


The $x$ variable represents the neural network's single input. The $w$ and $b$ variables specify the neural network's weight and bias. The equation above combines the weighted sum of the inputs with the sigmoid activation function. For this section, we will consider the sigmoid function because it demonstrates the effect of a bias neuron.


The weights of the neuron allow you to adjust the slope or shape of the activation function. Figure 3.A-WEIGHT shows the effect on the output of the sigmoid activation function if the weight is varied:

**Figure 3.A-WEIGHT: Neuron Weight Shifting**
![Adjusting Weight](https://data.heatonresearch.com/images/wustl/class/class_2_bias_weight.png "Neuron Weight Shifting")

The above diagram shows several sigmoid curves using the following parameters:

$$ f(x,0.5,0.0) $$
$$ f(x,1.0,0.0) $$
$$ f(x,1.5,0.0) $$
$$ f(x,2.0,0.0) $$

We did not introduce bias to produce the curves, as evidenced by the third parameter being 0 in each case. Using four weight values yields four distinct sigmoid curves in the figure above. No matter the weight, we always get the same value of 0.5 when _x_ is 0 because all curves hit the same point when x is 0. We might need the neural network to produce other values when the input is near 0.5.

Bias shifts the sigmoid curve, allowing values other than 0.5 when _x_ is near 0. Figure 3.A-BIAS shows the effect of using a weight of 1.0 with several different biases:

**Figure 3.A-BIAS: Neuron Bias Shifting**
![Adjusting Bias](https://data.heatonresearch.com/images/wustl/class/class_2_bias_value.png "Neuron Bias Shifting")

The above diagram shows several sigmoid curves with the following parameters:


$$ f(x,1.0,1.0) $$
$$ f(x,1.0,0.5) $$
$$ f(x,1.0,1.5) $$
$$ f(x,1.0,2.0) $$

We used a weight of 1.0 for these curves in all cases. When we utilized several different biases, sigmoid curves shifted to the left or right. Because all the curves merge at the top-right or bottom-left, it is not a complete shift.

When we combined bias and weights, they produced a curve that generated the desired output. The above curves are the output from only one neuron. In a complete network, the output from many different neurons will combine to produce intricate output patterns.



## Activation Functions

Activation functions compute the output of each neuron given its weighted input sum. Without a nonlinear activation function, a neural network of any depth reduces to a single linear transformation. The choice of activation function affects training stability, gradient flow, and convergence speed.

Modern deep networks use a small set of activation functions. The appropriate choice depends on the layer's role:

* **ReLU (and variants)** — hidden layers in feedforward and convolutional networks
* **GELU** — hidden layers in transformers and most modern language models
* **Sigmoid** — binary classification output layers; gating in LSTMs
* **Tanh** — gating in LSTMs; output layers requiring the range $(-1, 1)$
* **Softmax** — multiclass classification output layers
* **Linear** — regression output layers

## Linear Activation Function

The linear activation function returns its input unchanged:

$$ \phi(x) = x $$

**Figure 3.LIN: Linear Activation Function**
![Linear Activation Function](https://data.heatonresearch.com/images/wustl/class/graphs-linear.png "Linear Activation Function")

Regression networks use a linear activation on the output layer so that predictions are unconstrained real values. It also appears in the output layer for two-class classification when using a threshold decision rule rather than a probability.

## Rectified Linear Unit (ReLU)

ReLU is the default activation function for hidden layers in feedforward and convolutional networks:

$$ \phi(x) = \max(0, x) $$

**Figure 3.RELU: ReLU Activation Function**
![Rectified Linear Units (ReLU)](https://data.heatonresearch.com/images/wustl/class/graphs-relu.png "Rectified Linear Units (ReLU)")

ReLU displaced sigmoid and tanh in hidden layers because its derivative does not saturate for positive inputs. The derivative is:

$$ \phi'(x) = \begin{cases} 1 & x > 0 \\ 0 & x \leq 0 \end{cases} $$

When sigmoid is used in a deep network, the derivative of the sigmoid—shown below—saturates toward zero for large positive or negative inputs. During backpropagation, multiplying many near-zero derivatives across layers causes gradients to vanish before reaching the early layers. This is the **vanishing gradient problem**, and it was a primary obstacle to training deep networks before ReLU.

$$ \phi'(x)=\phi(x)(1-\phi(x)) $$

**Figure 3.SDERV: Sigmoid Derivative**
![Sigmoid Derivative](https://data.heatonresearch.com/images/wustl/class/class_2_deriv_sigmoid.png "Sigmoid Derivative")

The sigmoid derivative peaks at 0.25 when $x = 0$ and falls toward zero in both directions. ReLU's derivative is exactly 1 for all positive inputs, which allows gradients to pass through deep networks without attenuation.

ReLU has one known failure mode: neurons whose input is persistently negative receive a zero gradient and stop updating entirely. This is called the **dying ReLU problem**. Variants such as Leaky ReLU and ELU address this by allowing a small nonzero output for negative inputs, though plain ReLU remains the standard for most convolutional architectures.

## GELU

The Gaussian Error Linear Unit (GELU) is the activation function used in transformers, including GPT and BERT, and has largely replaced ReLU in language model architectures. [[Cite:hendrycks2016gelu]](https://arxiv.org/abs/1606.08415)

$$ \phi(x) = x \cdot \Phi(x) $$

where $\Phi(x)$ is the cumulative distribution function of the standard normal distribution. In practice, it is approximated as:

$$ \phi(x) \approx 0.5x\left(1 + \tanh\!\left[\sqrt{\frac{2}{\pi}}\left(x + 0.044715x^3\right)\right]\right) $$

Unlike ReLU, GELU is smooth and nonmonotonic near zero: it allows small negative outputs for inputs slightly below zero rather than clamping to zero. This smoothness stabilizes training in deep transformer networks. When working with transformer architectures later in this course, GELU is the activation function you will encounter in the feed-forward sublayers.

GELU has a formal connection to dropout. The function can be derived by considering a stochastic process where each input is randomly set to zero with a probability that depends on its magnitude — essentially a form of adaptive dropout applied at the neuron level. The GELU output $x \cdot \Phi(x)$ is the expected value of that process: $\Phi(x)$ is the probability that the input is kept, so larger positive values are almost always passed through while values near zero are suppressed with probability close to 0.5. Where ReLU gates inputs deterministically by sign, GELU gates them probabilistically by magnitude. This gives GELU a smoother gradient near zero and an implicit regularization effect that partly explains its stability in very deep transformer networks.

Regularization refers to any technique that reduces a model's tendency to overfit — to memorize the training data rather than learn generalizable patterns. Dropout is one of the most common regularization techniques for neural networks: during training, each neuron is randomly set to zero with some probability $p$ (typically 0.1 to 0.5) on each forward pass, forcing the network to learn redundant representations that do not rely on any single neuron. At inference, dropout is disabled and all neurons are active, with their outputs scaled to compensate. GELU's connection to dropout is that its shape arises from the same probabilistic gating idea, but applied deterministically as a fixed function rather than as a random process toggled on and off between training and inference. Dropout is covered in detail later in this book; the connection to GELU is worth revisiting once that material is complete.

When deciding between ReLU and GELU consider:

* **Transformers and attention-based models**: GELU is standard. This is where you'll see it in PyTorch and HuggingFace defaults.
* **CNNs and feedforward networks on tabular/image data**: ReLU remains the default and is still what most current research uses. ResNets, EfficientNets, and similar architectures still ship with ReLU.

## Sigmoid

The sigmoid function maps any real input to the range $(0, 1)$:

$$ \phi(x) = \frac{1}{1 + e^{-x}} $$

**Figure 3.SIGMOID: Sigmoid Activation Function**
![Sigmoid Activation Function](https://data.heatonresearch.com/images/wustl/class/graphs-sigmoid.png "Sigmoid Activation Function")

Sigmoid is no longer used in hidden layers of deep networks — the vanishing gradient problem described in the ReLU section applies directly to sigmoid. It remains relevant in two specific roles: the output layer of a binary classifier, where the output is interpreted as $P(y=1 \mid x)$, and the input and forget gates of LSTMs, where a value in $(0, 1)$ naturally represents how much information to pass or discard.

## Hyperbolic Tangent (Tanh)

The hyperbolic tangent function maps inputs to the range $(-1, 1)$:

$$ \phi(x) = \tanh(x) $$

**Figure 3.HTAN: Hyperbolic Tangent Activation Function**
![Hyperbolic Tangent Activation Function](https://data.heatonresearch.com/images/wustl/class/graphs-tanh.png "Hyperbolic Tangent Activation Function")

Tanh has the same shape as sigmoid but is zero-centered, which improves gradient flow compared to sigmoid in shallow networks. Like sigmoid, it saturates for large inputs and is not used in deep hidden layers. Its primary current role is in the cell output of LSTMs, where the output gate applies tanh to the cell state before passing it forward.

## Softmax

Softmax is used in the output layer of multiclass classifiers. It converts a vector of raw scores into a probability distribution over $k$ classes:

$$ \phi_i(\mathbf{x}) = \frac{\exp(x_i)}{\sum_{j=1}^{k} \exp(x_j)} $$

The outputs are all positive and sum to 1, so each output can be interpreted as the probability that the input belongs to class $i$. Unlike every other activation function in this module, the softmax output at position $i$ depends on the values at all other positions.

To see softmax concretely, consider a three-class classifier outputting the raw vector $[0.9, 0.2, 0.4]$. These values sum to 1.5 and cannot be treated as probabilities directly. Applying softmax:

$$ \text{sum} = \exp(0.9) + \exp(0.2) + \exp(0.4) = 5.1728 $$

$$ \phi_0 = \exp(0.9) / \text{sum} = 0.4755 $$
$$ \phi_1 = \exp(0.2) / \text{sum} = 0.2361 $$
$$ \phi_2 = \exp(0.4) / \text{sum} = 0.2884 $$

The result $[0.4755, 0.2361, 0.2884]$ sums to 1.0. The network assigns the highest probability (47.6%) to class 0.

## Step Activation Function

The step function is the activation used in the original perceptron, introduced by McCulloch & Pitts (1943). [[Cite:mcculloch1943logical]](https://link.springer.com/article/10.1007/BF02478259) It returns 1 if the input meets a threshold and 0 otherwise:

$$ \phi(x) = \begin{cases} 1 & x \geq 0.5 \\ 0 & x < 0.5 \end{cases} $$

**Figure 3.STEP: Step Activation Function**
![Step Activation Function](https://data.heatonresearch.com/images/wustl/class/graphs-step.png "Step Activation Function")

The step function has no well-defined derivative at the threshold and a zero derivative everywhere else, making it incompatible with gradient-based training. It is covered here for historical context only; no modern network uses it.

# GELU vs RELU

ReLU and GELU produce similar outputs for most inputs, but differ in one important region. ReLU clamps all negative inputs to exactly zero. GELU allows a small negative output for inputs near zero — the function dips to approximately $-0.17$ at $x \approx -0.17$ before returning to zero as $x \to -\infty$. This means GELU never fully suppresses a neuron based on sign alone; inputs slightly below zero still carry a weak negative signal through.

This distinction matters in transformers, where the smooth gradient near zero stabilizes training across very deep stacks of attention and feed-forward layers. For the feedforward networks covered in this module, ReLU remains the appropriate default. The following code plots both functions for comparison.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

def gelu(x):
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))

x = np.linspace(-4, 4, 1000)

plt.figure(figsize=(6, 3))
plt.plot(x, gelu(x), label='GELU')
plt.plot(x, np.maximum(0, x), label='ReLU', linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.ylim(-0.5, 3)
plt.legend()
plt.title('GELU vs ReLU')
plt.tight_layout()
plt.savefig('graphs-gelu.png', dpi=150)
plt.show()
```
